```python
# 나라 이름 넣기
import csv

keys = ['seoul','beijing', 'binzhou', 'dezhou', 'dongying', 'heze', 'jinan', 'jining', 'liaocheng', 'linyi', 'qingdao', 'rizhao', 'shanghai', 'taian', 'weifang', 'weihai', 'yantai', 'zaozhuang', 'zibo']
# keys 19개
for i in keys:
    with open(str(i) + '-air-quality.csv', 'r', encoding='UTF8') as file:
        reader = csv.reader(file)
        all = []
        row = next(reader)
        row.append('country')
        all.append(row)
        
        for row in reader:
            row.append(str(i))
            all.append(row)
            
    
    with open(str(i) + '-air-quality.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(all)
        file.close()
        
```

```python
# 결측값 처리
import csv

keys = ['seoul','beijing', 'binzhou', 'dezhou', 'dongying', 'heze', 'jinan', 'jining', 'liaocheng', 'linyi', 'qingdao', 'rizhao', 'shanghai', 'taian', 'weifang', 'weihai', 'yantai', 'zaozhuang', 'zibo']
# keys 19개
for i in keys:
    with open(str(i) + '-air-quality.csv', 'r', encoding='UTF-8') as file:
        reader = csv.reader(file)
        list = []
        all = []
        for text in reader:
            list.append(text)
        for j in range(len(list)):
            if list[j][1] == " ":
                list[j][1] = list[j-1][1]
            if list[j][2] == " ":
                list[j][2] = list[j-1][2]
            all.append(list)
#         print(list)
    
    with open(str(i) + '-air-quality.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(list)
        file.close()
```

```python
# 날짜 전처리
import csv
import re

keys = ['seoul','beijing', 'binzhou', 'dezhou', 'dongying', 'heze', 'jinan', 'jining', 'liaocheng', 'linyi', 'qingdao', 'rizhao', 'shanghai', 'taian', 'weifang', 'weihai', 'yantai', 'zaozhuang', 'zibo']
# keys 19개
for i in keys:
    with open(str(i) + '-air-quality.csv', 'r', encoding='UTF-8') as file:
        reader = csv.reader(file)
        lines = []
        for line in reader:
            line[0] = re.sub('/', '', line[0])
            lines.append(line)
        print(lines)
```
